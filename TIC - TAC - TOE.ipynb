```python
from IPython.display import clear_output
board = [" "," "," "," "," "," "," "," "," "]

def tic(pos,sym):
    if board[pos-1] == ' ':
        board[pos-1] = sym
        clear_output()
        print(" " + board[6] + " " + "|" + " " + board[7] + " " + "|" + " " + board[8] + " ")
        print("-----------")
        print(" " + board[3] + " " + "|" + " " + board[4] + " " + "|" + " " + board[5] + " ")
        print("-----------")
        print(" " + board[0] + " " + "|" + " " + board[1] + " " + "|" + " " + board[2] + " ")
        return "\n*** Updated Board ***"
    else:
        again(sym)
        return "" 

def wrong(sym):
    print("\n\t\t\t*** Wrong Input ***")
    if sym == 'X':
        x = int(input("\n\t\t\tPlayer 1 Please Enter again a Vald Position b/w 1-9: "))
        if x>9 or x<1:
            clear_output()
            x = wrong(sym)
        return x
    else:
        x = int(input("\n\t\t\tPlayer 2 Please Enter again a Vald Position b/w 1-9: "))
        if x>9 or x<1:
            clear_output()
            x = wrong(sym)
        return x

def again(sym):
    
    print("\n\t\t\t*** Position alredy holds a symbol ***")
    if sym == 'X':
        x = int(input("\n\t\t\tPlayer 1 Please Enter again a Vald Position : "))
        if board[x-1] != ' ':
            again(sym)
        else:
            print(tic(x,sym))
    else:
        x = int(input("\n\t\t\tPlayer 2 Please Enter again a Vald Position : "))
        if board[x-1] != ' ':
            again(sym)
        else:
            print(tic(x,sym))
    
def check(sym):
    for i in [0,1,2]:
        if board[i] == sym and board[i+3] == sym and board[i+6] == sym:
            return False
    for i in [0,3,6]:
        if board[i] == sym and board[i+1] == sym and board[i+2] == sym:
            return False
    if board[0] == sym and board[4] == sym and board[8] == sym:
            return False
    if board[2] == sym and board[4] == sym and board[6] == sym:
            return False
    
print("\n\t\t\tENJOY PLAYING TIC - TAC - TOE # GAME")
print("\t\t\t....................................")
print("\n\tThe 1st player gets the symbol 'X' and the 2nd player gets the symbol 'O' in the game\n")

cond = True
s = " "
x = 0
i = 0
while True:
    a = int(input("\n\t\t\t1st Player, Enter the Position for your 'X': "))
    if a>9 or a<1:
        a = wrong('X')
    print(tic(a,'X'))
    cond = check('X')
    if cond == False:
        s = 'X'
        break
    i+=1
    if i==9:
        break
    b = int(input("\n\t\t\t2nd Player, Enter the Position for your 'O': "))
    if b>9 or b<1:
        b = wrong('O')
    print(tic(b,'O'))
    cond = check('O')
    if cond == False:
        s = 'O'
        break
    i+=1
        
if i != 9:
    if s == 'X':
        print("\n\t\t\t..... Congratulations, Player 1 - THE WINNER .....")
    elif s == 'O':
        print("\n\t\t\t..... Congratulations, Player 2 - THE WINNER .....")
else:
    print("\n\t\t\t..... Congratulations, NO ONE WON, NO ONE LOST .....")
```
